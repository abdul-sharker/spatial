```python
import geopandas as gpd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Read shapefile
bd = gpd.read_file("C:/Users/hp/Desktop/Ratargul Map/bgd_adm_bbs_20201113_shp/bgd_adm_bbs_20201113_SHP/bgd_admbnda_adm0_bbs_20201113.shp")

# Set up the projection
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'projection': ccrs.PlateCarree()})
ax.set_extent([87, 93, 20, 27])  # Rough bounding box of Bangladesh

# Add country border
bd.boundary.plot(ax=ax, edgecolor='black', linewidth=1)

# Add gridlines with lat/lon
gl = ax.gridlines(draw_labels=True, linewidth=0.5, color='gray', alpha=0.5, linestyle='--')
gl.top_labels = gl.right_labels = False

# Add scale bar manually (simple example)
from matplotlib_scalebar.scalebar import ScaleBar
scalebar = ScaleBar(1, units='km', dimension='si-length')  # 1:1 scale (you can adjust this)
plt.gca().add_artist(scalebar)

# Add title
plt.title("Map of Bangladesh with Latitude & Longitude Grid")

plt.show()
```

```python
import matplotlib.pyplot as plt
import geopandas as gpd
import cartopy.crs as ccrs

# Load shapefile
bd = gpd.read_file("C:/Users/hp/Desktop/Ratargul Map/bgd_adm_bbs_20201113_shp/bgd_adm_bbs_20201113_SHP/bgd_admbnda_adm0_bbs_20201113.shp")

# Create figure
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'projection': ccrs.PlateCarree()})
bd.boundary.plot(ax=ax, edgecolor='black')
ax.set_extent([87, 93, 20, 27])

# Add North arrow
def add_north_arrow(ax, x=0.9, y=0.1, size=20):
    ax.text(x, y, 'N', transform=ax.transAxes,
            fontsize=size, fontweight='bold', ha='center', va='center')
    ax.arrow(x, y-0.03, 0, 0.05, transform=ax.transAxes,
             linewidth=2, head_width=0.02, head_length=0.02,
             fc='k', ec='k')

add_north_arrow(ax)

# Add gridlines
gl = ax.gridlines(draw_labels=True, linewidth=0.5, color='gray', linestyle='--')
gl.top_labels = gl.right_labels = False

plt.title("Map of Bangladesh with North Arrow", fontsize=15)
plt.savefig("bangladesh_with_north_arrow.png", dpi=300)
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import matplotlib.cm as cm

# Load ADM1 (division) level shapefile
bd = gpd.read_file("C:/Users/hp/Desktop/Ratargul Map/bgd_adm_bbs_20201113_shp/bgd_adm_bbs_20201113_SHP/bgd_admbnda_adm1_bbs_20201113.shp")

# Create figure
fig, ax = plt.subplots(figsize=(12, 12), subplot_kw={'projection': ccrs.PlateCarree()})
ax.set_extent([87, 93, 20, 27])

# Plot each division with a different color
bd.plot(ax=ax, column='ADM1_EN', cmap='Set3', legend=True, edgecolor='black')

# Add North Arrow
def add_north_arrow(ax, x=0.1, y=0.9, size=20):
    ax.text(x, y, 'N', transform=ax.transAxes, fontsize=size, fontweight='bold', ha='center', va='center')
    ax.arrow(x, y - 0.03, 0, 0.05, transform=ax.transAxes,
             linewidth=2, head_width=0.02, head_length=0.02, fc='k', ec='k')
add_north_arrow(ax)

# Add gridlines
gl = ax.gridlines(draw_labels=True, linewidth=0.5, color='gray', linestyle='--')
gl.top_labels = gl.right_labels = False

plt.title("Colorful Map of Bangladesh Divisions with North Arrow", fontsize=15)

plt.show()
```
